```python
q1=df['median_house_value'].quantile(0.25)
q3=df['median_house_value'].quantile(0.75)
iqr=q3-q1

lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr

df=df[(df['median_house_value']>=lower_bound)&(df['median_house_value']<=upper_bound)]
```
